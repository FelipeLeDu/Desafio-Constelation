```python
#variação dólar

var = ["USDBRL=X"]
start_date = "1990-01-01"
end_date = "2023-12-02"

data = yf.download(var, start=start_date, end=end_date)

data['Variacao_Percentual'] = (data['Close'] - data['Open']) / data['Open'] * 100
data.reset_index(inplace=True)
data = data[["Variacao_Percentual","Date"]]
data
```
